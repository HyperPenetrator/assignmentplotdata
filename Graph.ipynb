```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\LENOVO\Downloads\heart_disease_uci.csv")

plt.figure(figsize=(6,4))
df['sex'].value_counts().plot(kind='bar',
color=['green','red'])
plt.title("Distribution of Patients by Gender")
plt.xlabel("Gender")
plt.ylabel("Cholesterol")
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(df['age'], df['chol'],alpha=0.6, color='green')
plt.title("Age vs Cholestrol Levels as recorded")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()

plt.figure(figsize=(8,5))
df['age'].value_counts().plot(kind='bar',
color=['skyblue','yellow'])
plt.title("Maximum Heart Rate vs Heart Disease Presence")
plt.xlabel("Ages in which Heart disease is seen(Gradient)")
plt.ylabel("Maximum Heart Rate (thalach)")
plt.show()
```
